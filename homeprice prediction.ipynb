**2.Import and read the csv file**

```
import pandas as pd
df=pd.read_csv('/content/gdrive/MyDrive/DIU/ML practice/01. Linear Regression/2. Linear Regression Multiple Variables/1. Homeprice/homeprices.csv')
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

**3. Ploting attributes in a scatter plot**

```
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

**4. Fitting and training the model**




```
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```
reg.score(df[['area']],df.price)
```

```
reg.predict([[5800]])
```

```
reg.coef_
```

```
reg.intercept_
```

**5.Ploting the Linear regression plot**





```
plt.xlabel('area(sqr ft)', fontsize=20)
plt.ylabel('price(US$)',fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
